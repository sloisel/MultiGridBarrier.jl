# MultiGridBarrier.jl

**Author: Sébastien Loisel**

`MultiGridBarrier.jl` is a Julia package that solves convex variational problems, e.g. nonlinear PDEs and BVPs, using the MultiGrid Barrier method (with either finite elements or spectral elements), which is theoretically optimal for some problem classes.

See the documentation above for more information and examples. The rest of this document is an IPython Notebook that features a few simple example. 

## Quick tutorial.

First, one should add the package `MultiGridBarrier`.

```julia
using Pkg
Pkg.add("MultiGridBarrier")
```

### Solve some $p$-Laplace problems.

We begin with the case $p=1$ in 1d, using piecewise linear elements.

```julia
using MultiGridBarrier
fem1d_solve(L=5,p=1.0);
```

### Solve a 2d problem, piecewise quadratic elements.
$p$-Laplace problem with $p=1$.
The elements also include bubble functions so that the mass matrix is positive definite.

```julia
using MultiGridBarrier
fem2d_solve(L=3,p=1.0);
```

### Solve a 1d problem using spectral elements.
Here, $p=1$ and there are $40$ interpolation nodes, so the polynomial degree is $39$.

```julia
spectral1d_solve(n=40,p=1.0);
```

### 2d problem, spectral elements.
Solve a 2d $p$-Laplacian, with $p=1.5$.
There are $n=5$ interpolation nodes along each axis (so $25$ interpolation nodes in total.

```julia
spectral2d_solve(n=5,p=1.5);
```

# Time-dependent problem
All the above problems were spatial only (i.e. "elliptic"), now we solve a time-dependent problem with two spatial dimensions using piecewise quadratic elements (and bubble functions).

```julia
parabolic_solve(h=0.1,L=3);
```
